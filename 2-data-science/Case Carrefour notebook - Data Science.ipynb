## Apriori

O modelo de ML Apriori trabalha com a Regra de Associação, que nada mais é que um método de explorar relações entre Itens em conjuntos de dados. Principalemnte focado em cesta de supermercado, este modelo precisa trabalhar com uma base onde os itens aparecem de forma binaria para poder trazer a maior correlação entre eles. 

Vale ressaltar algumas das métricas que este modelo trabalha:
 - Support = Transações com o item / Total transações

 - Confidence = (Pão -> Leite) = Transações contendo pão e leite / Transaçõs contendo pão

 - Lift = Confidence (Pão -> Leite) / Support(Pão) = 0.75/1 = 1.3.

```python
#Foi selecionado uma amostra de 10.000 linhas da base para podermos rodar o modelo
#dados_bazar = dados[~dados['item_descricao'].str.startswith('SACOLA PLASTICA')].head(10000).copy()
dados_bazar = dados.head(10000).copy()
dados_bazar
```

```python
dados_bazar.shape[0]
```

```python
# Adequa a base para usar o algoritimo 
basket = pd.get_dummies(dados_bazar['item_descricao'])

basket['ticket'] = dados_bazar['ticket']

# Agrupar os itens por tickets
basket = basket.groupby('ticket').sum()

# Converter as contagens em 0s e 1s indicando se um item foi comprado ou não
basket_sets = basket.applymap(lambda x: 1 if x > 0 else 0)

# Exibir o DataFrame preparado
print(basket_sets)
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# trazer o calculo de support dos itens (support minimo de 0.01 por conta do volume de dados)
frequent_itemsets = apriori(df = basket_sets, min_support=0.01, use_colnames=True)
frequent_itemsets
```

```python
# Adicionar os demais calculos (confidence e lift) assim podemos ver a relação dos itens
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.01)
rules
```

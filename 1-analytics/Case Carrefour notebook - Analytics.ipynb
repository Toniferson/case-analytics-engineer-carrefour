## Top 10 Itens mais vendidos - Quantidade

```python
#quantidade de itens vendidos durante o periodo
dados_itens = dados[['item_descricao','item_unidade']].groupby('item_descricao')['item_unidade'].sum().sort_values(ascending = False)
dados_itens.head(10)
```

```python
#pode-se perceber que alguns dos itens que mais sao comercializados sao sacolas plasticas que nao acabam nao fazendo parte da cesta de compras dos clientes, entao serao excluidas da base para nao distorcer nossas analises
#excluindo item_descricao que comeca com sacola
dados_sem_sacola = dados[~dados['item_descricao'].str.startswith('SACOLA PLASTICA')]

#fazendo top 10 itens da base sem considerar as sacolas plasticas\
top10_sem_sacola = dados_sem_sacola[['item_descricao','item_unidade']].groupby('item_descricao')['item_unidade'].sum().sort_values(ascending = False).head(10)
top10_sem_sacola
```

## Top 10 Itens mais relevantes - Valor

```python
#foi feito o top 10 itens mais vendidos porem, como precisamos entender o top 10 itens precisamos agora cruzar com o valor do preço do item para podermos entender qual a representatividade dos produtos
dados_itens_preco = dados[['item_descricao','receita_bruta']].groupby('item_descricao')['receita_bruta'].sum().sort_values(ascending = False).head(10)
dados_itens_preco = pd.DataFrame(dados_itens_preco)
dados_itens_preco = dados_itens_preco.reset_index()
```

## Top 15 itens que acompanham os mais vendidos

```python
top10 = dados_itens_preco 
```

```python
#sera nescessario criar um df com todos os itens e todos os tickets
dados_ticket = dados[['item_descricao','ticket']]
dados_ticket
```

```python
tickets = pd.merge(dados_itens_preco,dados_ticket, on = 'item_descricao')
tickets = tickets[['item_descricao','ticket']]
tickets
```

```python
lista_ticket = tickets['ticket'].unique().tolist()
lista_ticket
```

```python
len(tickets)
```

```python
print(dados_ticket)
```

```python
df_final = pd.DataFrame()
for item in top10.itertuples():

    dados_ticket_filtrado = dados_ticket.loc[dados_ticket['item_descricao'] == item.item_descricao]
    lista_ticket = dados_ticket_filtrado['ticket'].unique().tolist()

    base_completa_aux = dados.copy()
    base_completa_aux = base_completa_aux[base_completa_aux['ticket'].isin(lista_ticket)]

    base_completa_aux = base_completa_aux.loc[base_completa_aux['item_descricao'] != item.item_descricao]
    base_completa_aux['item_top10'] = item.item_descricao

    #caso seja nescessario fazer um groupby pelo py, rodar os seguintes codigos
    #base_completa_aux = base_completa_aux.groupby('item_descricao')['receita_bruta'].count().sort_values(ascending = False).head(15)
    #base_completa_aux = base_completa_aux.groupby('item_descricao')['receita_bruta'].sum().sort_values(ascending = False).head(15)
    df_final = pd.concat([df_final, base_completa_aux], ignore_index=True) 
```
